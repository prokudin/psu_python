```python
def isWordGuessed(secretWord, lettersGuessed):
    '''
    secretWord: string, the word the user is guessing
    lettersGuessed: list, what letters have been guessed so far
    returns: boolean, True if all the letters of secretWord are in lettersGuessed;
    False otherwise
    '''
    return all(letters in ''.join(lettersGuessed) for letters in secretWord)
```

```python
def letterInWord(guess, secretWord):
    '''
    guess: user input, the letter that they guessed
    secretWord: string, the word the user is guessing
    returns: boolean, True if a letter is in the word; False otherwise
    '''
    gotOne = False
    
    if guess in secretWord:
        gotOne = True
    else:
        gotOne = False
        
    return gotOne
```

```python
def getGuessedWord(secretWord, lettersGuessed):
    '''
    secretWord: string, the word the user is guessing
    lettersGuessed: list, what letters have been guessed so far
    returns: string, comprised of letters and underscores that represents
      what letters in secretWord have been guessed so far.
    '''
    currentGuess=''
    
    for letter in secretWord:
        
        if letter in lettersGuessed:
            currentGuess += letter
            
        else:
            currentGuess += ' _ '
    
    return currentGuess
```

```python
def getAvailableLetters(lettersGuessed):
    '''
    lettersGuessed: list, what letters have been guessed so far
    returns: string, comprised of letters that represents what letters have not
      yet been guessed.
    '''    
    alphabet = ['a','b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
    
    for letter in lettersGuessed:
        
        if letter in alphabet:
            alphabet.remove(letter)
            
    return ''.join(alphabet)
```

```python
import random

def randomHangmanGuess (lettersGuessed):
    '''
    lettersGuessed: list, what letters have have been guessed so far
    getAvailableLetter: function, return string of letters not guessed yet
    returns: char, a single letter that has not been guessed yet
    '''
    
    availableLetters = getAvailableLetters(lettersGuessed)
    availableLetters = list(availableLetters)
    
    guess = random.choice(availableLetters)
    
    return guess
```

```python
def gotOne(secretWord, lettersGuessed):

    state = False
    counter = 0
    
    for space in getGuessedWord(secretWord, lettersGuessed):
        
        if space == '_':
            counter = counter
        elif space == ' ':
            counter = counter
        else:
            counter += 1
            
    if counter > 0:
        state = True
    
    return state
```

```python
allWords = []
import random

with open ('words_alpha.txt') as f:
    for line in f:
        allWords.append(line.rstrip())      

def hangmanAI(lettersGuessed, secretWord, allWords, wrong): 
              
    vowels = ['a', 'e', 'i', 'o', 'u']
    guess = ''
    possibleWords = []
    
    for word in allWords:
        
        if len(word) == len(secretWord):
            possibleWords.append(word)
               
    while wrong > 0:
            
        if not gotOne(secretWord, lettersGuessed):   
            guess = random.choice(vowels)
            
            while guess in lettersGuessed:
                guess = random.choice(vowels)
        
        if gotOne(secretWord, lettersGuessed):
            
            position = 0
            
            for space in getGuessedWord(secretWord, lettersGuessed):
                
                if space == secretWord[position : position + 1]:
                    
                    for word in possibleWords:
                        if word[position : position + 1] != space:
                            possibleWords.remove(word)
                    position += 1
                    
                elif space == ' ':
                    position = position
                    
                else:
                    position += 1
                            
            newSetLetters = []   
            
            for word in possibleWords:
                for letter in word:
                    newSetLetters.append(letter)
                    
            guess = random.choice(newSetLetters)
        
            while guess in lettersGuessed:
                guess = random.choice(newSetLetters)
                
        return guess
```

```python
def hangman(secretWord):
    '''
    secretWord: string, the secret word to guess.

    Starts up an interactive game of Hangman.

    * At the start of the game, let the user know how many 
      letters the secretWord contains.

    * Ask the user to supply one guess (i.e. letter) per round.

    * The user should receive feedback immediately after each guess 
      about whether their guess appears in the computers word.

    * After each round, you should also display to the user the 
      partially guessed word so far, as well as letters that the 
      user has not yet guessed.

    Follows the other limitations detailed in the problem write-up.
    '''
    print('Welcome to Hangman!\nI am thinking of a word that is ' + str(len(secretWord)) + ' letters long.')
    
    wrong = 8
    lettersGuessed = []
    state = False
    
    while not state or wrong > 0:
        
        if wrong > 0:
            
            print('----------------------\nYou have ' + str(wrong) + ' guesses left.')
            print('letters available:' + getAvailableLetters(lettersGuessed))
            
            #guess = raw_input('Please enter a guess: ') #use this on for interactive game
            #guess = randomHangmanGuess(lettersGuessed)  #use this for random guesser
            guess = hangmanAI(lettersGuessed, secretWord, allWords, wrong)   # use this for AI 
            guess = guess.lower()
            print('Please enter a guess: ' + guess)     #use this for random guesser and AI
            
        else:
            
            print('----GAME OVER----\nYou ran out of guesses. The secret word was "' + secretWord + '"')
            break
        
        
        if guess in lettersGuessed:
            
            print('Oops! You already guessed "' + guess + '", try another letter.')
            print(getGuessedWord(secretWord, lettersGuessed))
            
        else:
            
            lettersGuessed.append(guess)
            state = isWordGuessed(secretWord, lettersGuessed)
            
            if state:
                
                print(getGuessedWord(secretWord, lettersGuessed))
                print('Congratulations, you won! The secret word was "' + secretWord +'"')
                
                break
                
            elif letterInWord(guess, secretWord) == False:
                
                print('Sorry "' + guess + '" is not in the secret word.')
                print(getGuessedWord(secretWord, lettersGuessed))
                wrong -= 1
                
            elif letterInWord(guess, secretWord) == True:
                
                print('Good one! ' + getGuessedWord(secretWord, lettersGuessed))
```
