```python
def is_word_guessed(secret_word, letters_guessed):
     
    '''
     secret_word = the hangman word
     letters_guessed = letters guessed by user
     output: True or False: whether or not the user guessed all the letters in word
     
    '''
     
    for every_letter in secret_word:
        #if every_letter is not in the letters guessed, then hangman solution is false
        if every_letter not in letters_guessed:
            return(False)
    return(True)


def get_word_guessed(secret_word, letters_guessed):
     
    '''
     secret_word = the hangman word
     letters_guessed = letters guessed by user
     output: the word with the guessed letters inputed and the not guessed letters a mystery
     
    '''
    #initialize place to put output
    progress_of_game = ""
    
    #for loop first loops through the first letter in secret word and compares it to the inputed letters
    #if the letters match, it adds it to the first sequence of the progress_of_game string, if not, the else 
    #statement adds a dashed line. 
    
    for any_letters in secret_word:
        #build string to show progress
        if any_letters in letters_guessed:  
            progress_of_game += any_letters
        else:
            progress_of_game += "_ "
    return(progress_of_game)
    

import string
alphabet = string.ascii_lowercase

def get_available_letters(letters_guessed):
     
    '''

     letters_guessed = letters guessed by user
     output: letters that are still available to guess from alphabet
     
     
    '''
    #initialize place to put output
    letters_used = ""
    
    #for loop first loops through the first letter in letters_guessed and compares it to all letters in the alphabet
    #if the letters do not match, it adds it to the first sequence of the letters_used string, if not, the else 
    #statement adds a dashed line. 
    
    for any_letters in alphabet:
        #build string to show progress of letters available
        if any_letters in letters_guessed:  
            letters_used += " "
        else:
            letters_used += any_letters
    return(letters_used)
    

import random

#snippet of code used to locate file path
#os.path.abspath('words.txt')

WORDLIST_FILENAME = '/home/njb5305/psu_python/Week3/words.txt'

def loadWords():
    """
    Returns a list of valid words. Words are strings of lowercase letters.
    """
    print("Loading word list from file...")
    # inFile: file
    inFile = open(WORDLIST_FILENAME, 'r')
    # line: string
    line = inFile.readline()
    # wordlist: list of strings
    wordlist = line.split()
    print("  ", len(wordlist), "words loaded.")
    return wordlist

def chooseWord(wordlist):
    """
    wordlist (list): list of words (strings)

    Returns a word from wordlist at random
    """
    return random.choice(wordlist)


## Load the list of words into the variable wordlist
## so that it can be accessed from anywhere in the program
wordlist = loadWords()
secret_word = chooseWord(wordlist)

#dumbhangman code
def randomchoice():
    x = random.choice(alphabet)
    return(x)


#smarter hangman code
def letterchoices():
    """
    first chooses all words with the same amount of letters, then picks most popular
    letters from those words
    
    """
    
    #initialize string
    test11 = ""
    #convert to list for output
    test1 = list(test11)
    #intialize string
    wordlist_string = ""
    #convert to list for manipulation
    wordlist = list(wordlist_string)
    #open file
    wordstring = open(WORDLIST_FILENAME).read()
    #split every word in the string apart, and if the secret word legnth and word length
    #are the same, covert the word to list, and concatenate the letters together
    for word in wordstring.split():
        if len(secret_word) == len(word):
            convert = list(word)
            wordlist += convert
            
    #use below to check result
    #print(wordlist)
    
    #import counter and operator for sorting
    from collections import Counter
    import operator
    
    #letter_counts is a dictionary that logs the frequency of each letter to the letter
    letter_counts = (Counter(wordlist))

    #for ever letter in letter_counts, first find the max using itemgetter, then delete that
    #maximum, then create a list of letters from highest fequency to lowest
    for i in range (len(letter_counts)):
        test = max(letter_counts.iteritems(), key=operator.itemgetter(1))[0]
        del letter_counts[test]
        test1 += test
        
    #print(test1)

    return(test1)
```

```python
#Hangman Code smart iteration attempt


def hangman(secret_word):

    #set initial condition for alpha (win/loss tracker)
    alpha = False
    
    #import needed packages
    import random
    from operator import itemgetter
    
    WORDLIST_FILENAME = '/home/njb5305/psu_python/Week3/words.txt'


    ## Load the list of words into the variable wordlist
    ## so that it can be accessed from anywhere in the program
    wordlist = loadWords()
    secret_word = chooseWord(wordlist)

    #secret word
    #secret_word = 'apple'
    #number of guesses (dificulty level)
    number_of_guesses = 8

    
    #intro: 
    print('Hello user, are you ready to play Hangman?')
    #let user know number of letters
    print('The word contains ' + str(len(secret_word)) + ' letters')
    print('====================================')

    #initialize string
    letters_guessed = ''

    #initialize i to use in embedded while loop to switch from letter choices from list
    i = 0
    
    #create forever looping true while loop - like c-code
    while(1):
        
        #set beta equal to false at the start
        #beta used for guess count correction factor
        beta = False
        
        #show number of guesses
        print('You have ' + str(number_of_guesses) + ' guesses left')

        
        ################# Use for smarter hangman
        
        #while loop incremements through the letters in the list and then breaks to continue 
        #playing hangman
        while(1):
            i += 1 
            user_input = itemgetter(i)(letterchoices())
            break
            
        ################# Use for dumb hangman
        
        #user_input = randomchoice()
        
        #test if letters guessed was already guessed
        if user_input in letters_guessed:
            print ('You already guessed that letter')
            #if guessed already, add 1 to the number of guesses
            number_of_guesses = number_of_guesses+1
            #set beta equal to true as a flag to show number of guesses has been increased
            beta = True

        #concatenate letters guessed string
        letters_guessed += user_input

        #check if letter in word
        if user_input in secret_word:
            print ('Letter Guessed: ' + user_input)
            print('You got one: ' + get_word_guessed(secret_word, letters_guessed))
            #add guess for correct answer
            number_of_guesses = number_of_guesses+1
            
            #recieve flag and correct for extra added guess
            if beta == True:
                number_of_guesses = number_of_guesses-1
                beta = False
                
        #show user letter not in word
        else :
            print ('Letter Guessed: ' + user_input)
            print ("Try again: " + get_word_guessed(secret_word, letters_guessed))

        
        #check if lost
        if number_of_guesses == 1:
            alpha = True
            print ('You lose! The word is ' + str(secret_word) + '.')
            #return boolean value to use as hangman game interation checker
            return(alpha)
            break
        #check if won
        if is_word_guessed(secret_word, letters_guessed) == 1:
            print ('You won! The word is ' + str(secret_word) + '.')
            break

        #print letters available
        #subtract off one guess
        number_of_guesses = number_of_guesses-1
        print('Letters Available: '+ get_available_letters(user_input))
        print('====================================')
        
        
```
