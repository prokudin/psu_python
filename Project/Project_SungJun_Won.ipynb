```python
import pandas as pd
file_name = "1000.xlsx"

class Data(object):
    def __init__(self):
        """
        Initialize data as data achieved from excel file 1000.xlsx
        self.data = excel data determined by pandas read_excel method
        """
        self.data = pd.read_excel(file_name)
        
    def get_data(self):
        """
        used to safely access self.data outside the class
        """
        return self.data
    
    def replace_data(self, new_file_name):
        """
        used to exchange old data with a new data
        """
        self.data = pd.read_excel(new_file_name)
        return self.data
    
    def build_data_frame(self):
        """
        used to build data frame
        """
        data = self.get_data()
        df = pd.DataFrame(data)
        return df

class Dataframe(Data):
    def __init__(self,row,column):
        """
        row(integer) row number
        column(integer) column number
        """
        Data.__init__(self)
        self.row = row
        self.column = column
        self.dataframe = Data.build_data_frame(self)
        
    def get_row(self):
        """
        Used to safely access self.row outside of the class
        """
        return self.row
    
    def get_column(self):
        """
        Used to safely access self.column outside of the class
        """
        return self.column
    
    def get_dataframe(self):
        """
        Used to safely access self.dataframe outside of the class
        """
        return self.dataframe
    
    def find_row(self, row):
        """
        Used to find specific row in dataframe
        """
        dataframe = Data.build_data_frame(self)
        result = dataframe.iloc[row]
        return result
    
    def find_column(self, column):
        """
        Used to find specific column in dataframe
        """
        dataframe = Data.build_data_frame(self)
        result = dataframe.iloc[:,column]
        return result
```
